```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points = 10000

A = np.random.random(num_points) * 2*np.pi
S = np.random.random(num_points)
```

```python
x = np.cos(A) * S
y = np.sin(A) * S

plt.figure(figsize=(6, 6))
plt.scatter(x, y, marker='.', s=0.1)
```

```python
r = np.sqrt(S)
x = np.cos(A) * r
y = np.sin(A) * r

_x = np.linspace(0, 1, 20)
_zero = np.zeros(_x.shape[0])

plt.figure(figsize=(6, 6))
plt.scatter(r, S, marker='.', s=0.001)
for idx in range(_x.shape[0]):
    plt.plot([np.sqrt(_x[idx]), np.sqrt(_x[idx]), 0], [0, _x[idx],  _x[idx]], c='c')
    
plt.figure(figsize=(6, 6))
plt.scatter(x, y, marker='.', s=1)
```

```python
r = np.arccos(S) / (np.pi/2)
x = np.cos(A) * r
y = np.sin(A) * r

_x = np.linspace(0, 1, 20)
_zero = np.zeros(_x.shape[0])

plt.figure(figsize=(6, 6))
plt.scatter(r, S, marker='.', s=0.001)
for idx in range(_x.shape[0]):
    plt.plot([np.arccos(_x[idx]) / (np.pi/2), np.arccos(_x[idx]) / (np.pi/2), 0], [0, _x[idx],  _x[idx]], c='c')

plt.figure(figsize=(6, 6))
plt.scatter(x, y, marker='.', s=1)
```

```python
plt.figure(figsize=(6, 6))
r = np.sqrt(S)
plt.scatter(r, S, marker='.', s=0.001)
r = np.arccos(S) / (np.pi/2)
plt.scatter(r, S, marker='.', s=0.001)
r = np.arccos(S.max() - S) / (np.pi/2)
plt.scatter(r, S, marker='.', s=0.001)
```
